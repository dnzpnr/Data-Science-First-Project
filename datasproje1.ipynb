```python
import pandas as pd
```

```python
purchases = pd.read_csv('purchases.csv')
```

```python
purchases
```

```python
users = pd.read_csv('users.csv')
```

```python
users
```

```python
# 1. users ve purchases veri setlerinin okutulması ve
#veri setlerinin "uid" değişkenine göre inner join ile merge edilmesi.
```

```python
df = pd.merge(purchases, users, on = 'uid')
```

```python
df
```

```python
#2. Kendi yazacağın fonksiyon ile veri setinin genel özelliklerinin betimlenmesi.
```

```python
def helpers_function(x):
    print('Observations : ' + str(x.shape[0]))
    print('Variables : ' + str(x.shape[1]))
    df_cat = pd.DataFrame(df.select_dtypes(include = 'object').columns)
    print('cat_col : ' + str(df_cat.count()[0]))
    df_num = pd.DataFrame(df.select_dtypes(include = ['float64','int64']).columns)
    print('num_col : ' + str(df_num.count()[0]))
```

```python
helpers_function(df)
```

```python
# 3. Veri setinin Shape, Types, Head, Tail, NA, Quantiles 
#verilerini bize getiren fonksiyonu tanımlama.
```

```python
def my_Describe_Func(x):
    print('Shape: ' + str(x.shape))
    print('Types: ' + str(x.dtypes))
    print('Head : ' + str(x.head))
    print('Tail : ' + str(x.tail))
    print('NaN : ' + str(x.isnull()))
    price_Q1 = df['price'].quantile(0.25)
    price_Q3 = df['price'].quantile(0.75)
    price_Q2 = (price_Q1 + price_Q3)/2
    print('price_Q1 : ' + str(price_Q1))
    print('price_Q2 : ' + str(price_Q2))
    print('price_Q3 : ' + str(price_Q3))
    age_Q1 = df['age'].quantile(0.25)
    age_Q3 = df['age'].quantile(0.75)
    age_Q2 = (age_Q1 + age_Q3)/2
    print('age_Q1 : ' + str(age_Q1))
    print('age_Q2 : ' + str(age_Q2))
    print('age_Q3 : ' + str(age_Q3))
    print('NaN : ' + str(x.isnull()))
```

```python
my_Describe_Func(df);
```
